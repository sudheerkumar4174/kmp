```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x=df.iloc[:,:-1]
y=df.iloc[::,-1]
```

```python
# kmeans without feature selection
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score


kmeans = KMeans(n_clusters=3)
kmeans.fit(x)

print("Silhouette Coefficient:", silhouette_score(x, kmeans.labels_))
print("Calinski-Harabasz Score:", calinski_harabasz_score(x, kmeans.labels_))
print("Davies-Bouldin Score:", davies_bouldin_score(x, kmeans.labels_))
```

```python

# without feature selection
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward")
hc.fit(x)

print("Silhouette Coefficient:", silhouette_score(x, hc.labels_))
print("Calinski-Harabasz Score:", calinski_harabasz_score(x, hc.labels_))
print("Davies-Bouldin Score:", davies_bouldin_score(x, hc.labels_))
```

```python

# mean shift clustering without feature selection
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(x)

ms = MeanShift(bandwidth=bandwidth)
ms.fit(x)


print("Silhouette Coefficient:", silhouette_score(x, ms.labels_))
print("Calinski-Harabasz Score:", calinski_harabasz_score(x, ms.labels_))
print("Davies-Bouldin Score:", davies_bouldin_score(x, ms.labels_))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python


from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(x_train)
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)
pca_data = pca.transform(df)
print(pca_data.shape)
pca_data
```

```python
# kmeans with pca
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
sscaler = StandardScaler()
pca_kmeans2 = KMeans(n_clusters=3)
pipeline_pca = make_pipeline(sscaler, pca_kmeans2)
pipeline_pca.fit(pca_data)

labels_pca = pipeline_pca.predict(pca_data)
print("Silhouette Coefficient:", silhouette_score(pca_data, labels_pca))
print("Calinski-Harabasz Score:", calinski_harabasz_score(pca_data, labels_pca))
print("Davies-Bouldin Score:", davies_bouldin_score(pca_data, labels_pca))
```

```python


# Performing hierarchical clustering with PCA data
hc_pca = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward")
hc_pca.fit(pca_data)


print("Silhouette Coefficient:", silhouette_score(pca_data, hc_pca.labels_))
print("Calinski-Harabasz Score:", calinski_harabasz_score(pca_data, hc_pca.labels_))
print("Davies-Bouldin Score:", davies_bouldin_score(pca_data, hc_pca.labels_))
```

```python
# mean shift clustering with pca
from sklearn.cluster import MeanShift, estimate_bandwidth


bandwidth = estimate_bandwidth(pca_data, quantile=0.2, n_samples=500)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
labels_ms = ms.fit_predict(pca_data)

silhouette_avg_ms = silhouette_score(pca_data, labels_ms)
ch_score_ms = calinski_harabasz_score(pca_data, labels_ms)


db_score_ms = davies_bouldin_score(pca_data, labels_ms)


print("Silhouette Coefficient (MeanShift):", silhouette_avg_ms)
print("Calinski-Harabasz score (MeanShift):", ch_score_ms)
print("Davies-Bouldin score (MeanShift):", db_score_ms)
```
